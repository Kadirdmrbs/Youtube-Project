```python
categories = ['Premier League','#PremierLeague','Serie A','#SerieA','WWE','NBA','D√ºnya Kupasƒ±','Formula 1','Son Viraj','UEFA']
```

```python
df2 = df.copy()
```

```python
df2
```

```python
df2['video_title']
```

```python
def my_func(title):
    for cat in categories:
        if cat.lower() in str(title).lower():
            return cat
```

```python
df2['categories'] = df2['video_title'].apply(my_func)
```

```python
df2 = df2.fillna('other')
```
