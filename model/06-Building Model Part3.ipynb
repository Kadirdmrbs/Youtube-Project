```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('finall.csv',index_col=0)
```

```python
df
```

```python
df2 = pd.get_dummies(df['categories'])
df3 = pd.concat([df,df2],axis=1)
df4 = pd.get_dummies(df['year'])
df5 = pd.concat([df3,df4],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

## CREATE X and y
X = df5.drop(['video_id','video_title','upload_date','categories','view_count','year'],axis=1)
y = df5['view_count']

# TRAIN TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()

model.fit(X_train,y_train)
base_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(y_test,base_pred)
```

```python
np.sqrt(mean_squared_error(y_test,base_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model,filled=True,feature_names=X.columns);
```

```python
from sklearn.model_selection import GridSearchCV

model = DecisionTreeRegressor()
param_grid = {'max_depth':[2,3,4,5,6,7,8],
             'criterion':['squared_error','entropy'],
              'max_leaf_nodes':[2,3,4,5,6,7,8,11,15,20,30,40,42,50],
             }

grid = GridSearchCV(model,param_grid=param_grid,cv=10,verbose=1,scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_preds = grid.predict(X_test)
```

```python
mean_absolute_error(y_test,grid_preds)
```

```python
np.sqrt(mean_squared_error(y_test,grid_preds))
```

## Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

model = RandomForestRegressor()
param_grid = {'n_estimators':[5,10,20,30,50,64,100,128,150],
              'max_features': [2,3,4,5,6,7,8,9,10],
              'bootstrap' : [True,False],
              'oob_score' : [True,False]
             }
grid = GridSearchCV(model,param_grid=param_grid,cv=10,verbose=1,scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_preds = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,grid_preds)
```

```python
np.sqrt(mean_squared_error(y_test,grid_preds))
```

# AdaBoost

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV

model = AdaBoostRegressor()
param_grid = {'n_estimators':[5,10,20,30,50,64,100,128,150],
              'loss':['linear', 'square', 'exponential']
             }
grid = GridSearchCV(model,param_grid=param_grid,cv=10,verbose=1,scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_preds = grid.predict(X_test)
```

```python
mean_absolute_error(y_test,grid_preds)
```

```python
np.sqrt(mean_squared_error(y_test,grid_preds))
```

```python
feat_import = grid.best_estimator_.feature_importances_
```

```python
imp_feats = pd.DataFrame(index=X.columns,data=feat_import,columns=['Importance'])
```

```python
imp_feats.sort_values("Importance",ascending=False)
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=imp_feats.sort_values('Importance'),x=imp_feats.sort_values('Importance').index,y='Importance')
plt.xticks(rotation=90);
```
